```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
matplotlib.style.use('ggplot')
%pylab inline
```

# mu=0.01

```python
file_list = ['index_'+file+'.csv' for file in [
"message_2016-6-4_1212H.log",
"message_2016-6-4_1224H.log",
"message_2016-6-4_1230H.log",
"message_2016-6-4_1236H.log",
"message_2016-6-4_1242H.log",
]]
df_list = [pd.read_csv(f, delimiter=';', index_col=0) for f in file_list]
for i, df in enumerate(df_list):
    df.loc[:, 'start_ts'] = df['start_ts'].apply(lambda x: pd.to_datetime(x))
    df.set_index('start_ts', inplace=True)
    df['total_reward'] = df['tr_reward'] - df['op_cost']
    print file_list[i]
```

```python
ts_name = ['N='+str(N) for N in [1,2,3,4,5,]]
styles = ['-']*6
```

```python
start = pd.to_datetime("2014-09-27 0:00:00")
end = pd.to_datetime("2014-09-28 00:00:00")
delta = pd.Timedelta('2 seconds')
plt.figure(1)
ax = plt.subplot(111)
for i, df in enumerate(df_list):
    step = (df.index-df.index[0])/delta+1
    ts = df['total_reward'].cumsum()/step
    ts.name = ts_name[i]
    ts.plot(figsize=(20, 10), legend=True, ax=ax, style=styles[i])
```
